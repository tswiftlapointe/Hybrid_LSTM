### 1. Monthly streamflow volumes

Download DAILY 2020 Level Modified Flows from Bonneville Power Administration: https://www.bpa.gov/energy-and-services/power/historical-streamflow-data into the directory /Data/BPA/

We only need the file MCD6A_daily.xlsx, the daily inflows at Mica. Save as csv file.

The following code will convert this to monthly accumulated volumes and save the monthly volumes as .csv file.

```python
## load BPA streamflow data at Mica

filename_BPA = 'MCD6A_daily.csv'
data_BPA = pd.read_csv(dir_data + '/' + filename_BPA, header=0)
data_BPA = np.array(data_BPA)
print(data_BPA.shape)
```

```python
## cut to start 1981-01-01 and end 2017-12-31

start_ind = np.where(data_BPA[:,0] == '1981-01-01')[0][0]
end_ind = np.where(data_BPA[:,0] == '2018-01-01')[0][0]
data_BPA_cut = data_BPA[start_ind:end_ind,:]
print(data_BPA_cut.shape)
```

```python
## sum daily streamflow in each month

years_all = np.arange(1981,2018,1)
leap_years = [1984, 1988, 1992, 1996, 2000, 2004, 2008, 2012, 2016]

# starting index of each month
month_starts = [0, 31, 59, 90, 120, 151, 181, 212, 243, 273, 304, 334, 365]
month_starts_leap = [0, 31, 60, 91, 121, 152, 181, 213, 244, 274, 305, 335, 366]

# placeholder for month volumes
data_monthly_volumes = np.empty((len(years_all),12))

# loop through all years and calculate monthly volumes
ind_start = 0
for i in range(len(years_all)):
    year = years_all[i]
    if year in leap_years:
        year_length = 366
    else:
        year_length = 365

    # get full year data, Jan1-Dec31
    data_year = data_BPA_cut[ind_start:ind_start+year_length,1]

    # convert from ft^3/s (cfs) to m^3/s
    cfs_to_m3s = 0.0283168
    data_year = data_year*cfs_to_m3s

    # multiply by (3600s/h)*(24h/day)
    data_year_vol = data_year*3600*24

    # sum monthly volumes
    for n in range(12):
        first_month = month_starts[n]
        second_month = month_starts[n+1]
        data_monthly_volumes[i,n] = np.sum(data_year_vol[first_month:second_month])

    # change start index
    ind_start += year_length
# end loop through all years

print('Done.')
print(data_monthly_volumes.shape)
```

```python
### save data as csv file

## reshape data into 1D in time
## & create array of dates

data_monthly_volumes_1D = np.zeros((len(years_all)*12))
dates_1D = np.empty((len(years_all)*12), dtype='U100')

# loop through years
for y, year in enumerate(years_all):
    for m in range(12):
        # add year's data
        data_monthly_volumes_1D[y*12+m] = data_monthly_volumes[y,m]
        # create month-year string
        if m+1 < 10:
            dates_1D[y*12+m] = '0' + str(m+1) + '-' + str(year)
        else:
            dates_1D[y*12+m] = str(m+1) + '-' + str(year)
# end loop through years

## save as csv
with open(dir_data + '/BPA_monthly_volumes_1981_2017.csv','w',newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Date', 'Volume (m^3)'])
    for i in range(len(dates_1D)):
        row_data = [dates_1D[i], data_monthly_volumes_1D[i]]
        writer.writerow(row_data)
```

### 2. Monthly Averaged Minimum & Maximum Temperature

Download ERA5 hourly data on single levels from 1940 to present (Hersbach et al., 2020): https://cds.climate.copernicus.eu/datasets/reanalysis-era5-single-levels?tab=download; two separate datasets for the two variables
- Variables: '2m temperature' and 'total precipitation'
- Years: 1981 to 2017 (inclusive)
- Months: all
- Days: all;
- Times: 00:00, 06:00, 12:00, 18:00 (temperature); All (precipitation)
- Subregion extraction: I download the whole province of BC, latitudes 48 to 60 degN; longitudes -142 to -109 degE

Due to data request size limits in Copernicus, I had to download the precipitation data in 3 separate files, split up by year range: 1981-1993, 1994-2006, 2007-2017

**Note:** must regrid this ERA5 data to the grid of the ECMWF SEAS5 hindcasts, I used CDO to do this.

Final regridded datasets should be located in Data/ directory, called:

ERA5_T2m_1981_2017_6hourly_ecmwfgrid.nc

ERA5_Precip_1981_1993_hourly_ecmwfgrid.nc, ERA5_Precip_1994_2006_hourly_ecmwfgrid.nc, ERA5_Precip_2007_2017_hourly_ecmwfgrid.nc

Will also need the Mica catchment area outline file in Data/ directory, called:

Mica_basin_outline.csv

The following code will convert these to monthly averages/sums, extract data for the Mica catchment area only, and save the monthly data as a .csv file. This is done for temperature first, then for each precipitation file separately.

```python
### load ERA5 datasets

## Some of the code in this cell is based off of code by Sam Anderson, from the publication:
# Anderson, S., and Radic, V. (2022). Evaluation and interpretation of convolutional long short-term memory
#    networks for regional hydrological modelling. Hydrol. Earth Syst. Sci., 26, 795-825.

### Temperature ###
filename_T2m = 'ERA5_T2m_1981_2017_6hourly_ecmwfgrid.nc'
ERA5_T2m = Dataset(dir_data + '/' + filename_T2m)

# Extract variables
ERA5_T2m_lon = ERA5_T2m.variables['longitude'][:]  # longitude in degW
ERA5_T2m_lat = ERA5_T2m.variables['latitude'][:]   # latitude in degW
ERA5_T2m_time = ERA5_T2m.variables['valid_time'][:]     # datetime64[ns]
ERA5_T2m_hourly = ERA5_T2m.variables['t2m'][:]      # 2m temperature (K)

print("Longitude shape = ", ERA5_T2m_lon.shape)
print("Latitude shape = ", ERA5_T2m_lat.shape)

# Get min/max of each day
numperday = 4 # 6-hourly data has 4 time points per day
ERA5_Tavg_daily = np.zeros((int(len(ERA5_T2m_time)/numperday),len(ERA5_T2m_lat), len(ERA5_T2m_lon)))
ERA5_Tmax_daily = np.zeros((int(len(ERA5_T2m_time)/numperday),len(ERA5_T2m_lat), len(ERA5_T2m_lon)))
ERA5_Tmin_daily = np.zeros((int(len(ERA5_T2m_time)/numperday),len(ERA5_T2m_lat), len(ERA5_T2m_lon)))
for daynum in range(int(len(ERA5_T2m_time)/numperday)): #for each day
        inds = list(range(daynum*numperday,daynum*numperday+numperday))
        ERA5_Tavg_daily[daynum,:,:] = np.mean(ERA5_T2m_hourly[inds,:,:],axis=0)
        ERA5_Tmax_daily[daynum,:,:] = np.max(ERA5_T2m_hourly[inds,:,:],axis=0)
        ERA5_Tmin_daily[daynum,:,:] = np.min(ERA5_T2m_hourly[inds,:,:],axis=0)

print(ERA5_Tavg_daily.shape)
print(ERA5_Tmax_daily.shape)
print(ERA5_Tmin_daily.shape)
```

```python
### get Mica basin outline from csv file

## load basin outline data
basin_outline = pd.read_csv('Mica_basin_outline.csv', header=0)
basin_outline = np.asarray(basin_outline)
basin_outline_x = basin_outline[:,0]
basin_outline_y = basin_outline[:,1]

## convert to Shapely Polygon structure
polygon = Polygon(zip(basin_outline_x,basin_outline_y))
basin_x, basin_y = polygon.exterior.xy

## will need to convert to lat,lon degrees for later
# first, need to determine edge of each grid box
# determine distance between each grid box
xstep = (ERA5_T2m_lon[-1]-ERA5_T2m_lon[0])/len(ERA5_T2m_lon)
ystep = (ERA5_T2m_lat[-1]-ERA5_T2m_lat[0])/len(ERA5_T2m_lat)
# get lon,lat values of top left corner of each grid box
gridedgex = np.arange(ERA5_T2m_lon[0],ERA5_T2m_lon[-1],xstep)
gridedgey = np.arange(ERA5_T2m_lat[0],ERA5_T2m_lat[-1],ystep)
# get conversion
myProj = Proj("EPSG:32610")
LON, LAT = np.meshgrid(gridedgex, gridedgey)
UTM_lon, UTM_lat = myProj(LON,LAT)
```

```python
### plot some data with Mica catchment area
## plot Tmax of day 1 as example

ocean = cfeature.NaturalEarthFeature('physical', 'ocean', '10m', edgecolor='w', alpha=0.2, facecolor='none', linewidth=1)
borders = cfeature.NaturalEarthFeature('cultural', 'admin_0_boundary_lines_land', '10m', edgecolor='k', linewidth=1, facecolor='none')
coasts = cfeature.GSHHSFeature(scale='intermediate', levels=[1], facecolor='none', edgecolor='k', alpha=0.7, linewidth=1)
provinces = cfeature.NaturalEarthFeature('cultural', 'admin_1_states_provinces_lines', '50m', facecolor='none', edgecolor='k', linewidth=1)

fig = plt.figure(figsize=(8,4))
ax1 = plt.axes(projection=ccrs.PlateCarree(), aspect='auto')
ax1.add_feature(ocean)
ax1.add_feature(coasts)
ax1.add_feature(borders)
ax1.add_feature(provinces)
im = ax1.imshow(ERA5_Tmax_daily[0,:,:], aspect='auto', cmap='viridis',
                extent=[min(ERA5_T2m_lon),max(ERA5_T2m_lon),min(ERA5_T2m_lat),max(ERA5_T2m_lat)])
plt.colorbar(im, label='Temperature (K)')
ax1.plot(basin_x, basin_y, 'r-', linewidth=2, transform=ccrs.epsg(32610), label='Mica basin')  # Mica basin outline
gl = ax1.gridlines(draw_labels = True)
gl.xlines = False
gl.ylines = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
gl.xlocator = mticker.FixedLocator(np.arange(-150,100,2))
gl.ylocator = mticker.FixedLocator(np.arange(40,70,1))
gl.xlabel_style = {'size':12}
gl.ylabel_style = {'size':12}
gl.top_labels = False
gl.right_labels = False
plt.title('ERA5 Max Temperature 01/01/1981')
plt.legend()
plt.show()
```

```python
### convert to monthly data
## temperature: average over all days in month; precip: sum all days in month

years_all = np.arange(1981,2018,1)
leap_years = [1980, 1984, 1988, 1992, 1996, 2000, 2004, 2008, 2012, 2016, 2020]
days_in_months = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
days_in_months_leap = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

# placeholder arrays
ERA5_Tavg_monthly = np.zeros((len(years_all)*12,len(ERA5_T2m_lat),len(ERA5_T2m_lon)))
ERA5_Tmax_monthly = np.zeros((len(years_all)*12,len(ERA5_T2m_lat),len(ERA5_T2m_lon)))
ERA5_Tmin_monthly = np.zeros((len(years_all)*12,len(ERA5_T2m_lat),len(ERA5_T2m_lon)))
#ERA5_P_monthly = np.zeros((len(years_all)*12,len(ERA5_P_lat),len(ERA5_P_lon)))

# loop through years
year_counter = 0
for y, year in enumerate(years_all):
    # get year, month lengths based on leap years
    if year in leap_years:
        year_length = 366
        month_lengths = days_in_months_leap
    else:
        year_length = 365
        month_lengths = days_in_months

    # get data for selected year
    #data_P_year = ERA5_P_daily[year_counter:year_counter+year_length,:,:]
    data_Tavg_year = ERA5_Tavg_daily[year_counter:year_counter+year_length,:,:]
    data_Tmax_year = ERA5_Tmax_daily[year_counter:year_counter+year_length,:,:]
    data_Tmin_year = ERA5_Tmin_daily[year_counter:year_counter+year_length,:,:]

    # loop through months in selected year
    month_counter = 0
    for m in range(12):
        month_length = month_lengths[m]
        # get data for selected month
        #data_P_month = data_P_year[month_counter:month_counter+month_length,:,:]
        data_Tavg_month = data_Tavg_year[month_counter:month_counter+month_length,:,:]
        data_Tmax_month = data_Tmax_year[month_counter:month_counter+month_length,:,:]
        data_Tmin_month = data_Tmin_year[month_counter:month_counter+month_length,:,:]
        # average / sum across month
        ERA5_Tavg_monthly[(y*12)+m,:,:] = np.mean(data_Tavg_month, axis=0)
        ERA5_Tmax_monthly[(y*12)+m,:,:] = np.mean(data_Tmax_month, axis=0)
        ERA5_Tmin_monthly[(y*12)+m,:,:] = np.mean(data_Tmin_month, axis=0)
        #ERA5_P_monthly[(y*12)+m,:,:] = np.sum(data_P_month,axis=0)

        month_counter += month_length
    # end loop through months

    year_counter += year_length
# end loop through years

print(ERA5_Tavg_monthly.shape)
print(ERA5_Tmax_monthly.shape)
print(ERA5_Tmin_monthly.shape)
#print(ERA5_P_monthly.shape)
```

```python
### load ERA5 precipitation datasets one at a time and convert to monthly data

## If have sufficient space on computer, can load all at once and combine

### Precipitation 1 ###
filename_P1 = 'ERA5_Precip_1981_1993_hourly_ecmwfgrid.nc'
ERA5_P1 = Dataset(dir_data + '/' + filename_P1)

# Extract variables
ERA5_P_lon = ERA5_P1.variables['longitude'][:]  # longitude in degW
ERA5_P_lat = ERA5_P1.variables['latitude'][:]   # latitude in degW
ERA5_P1_time = ERA5_P1.variables['valid_time'][:]     # hours since 01/01/1900
ERA5_P1_hourly = ERA5_P1.variables['tp'][:]      # total precipitation (m)

print("Longitude shape = ", ERA5_P_lon.shape)
print("Latitude shape = ", ERA5_P_lat.shape)

## Convert to daily
# Sum all precip in each day
numperday = 24  # hourly data has 24 time points per day
ERA5_P1_daily = np.zeros((int(len(ERA5_P1_time)/numperday),len(ERA5_P_lat), len(ERA5_P_lon)))
for daynum in range(int(len(ERA5_P1_time)/numperday)):
    inds = list(range(daynum*numperday, daynum*numperday+numperday))
    ERA5_P1_daily[daynum,:,:] = np.sum(ERA5_P1_hourly[inds,:,:],axis=0)

print(ERA5_P1_daily.shape)

## Convert to monthly
years_subset1 = np.arange(1981,1994,1)
ERA5_P1_monthly = np.zeros((len(years_subset1)*12,len(ERA5_P_lat),len(ERA5_P_lon)))

# loop through years
year_counter = 0
for y, year in enumerate(years_subset1):
    # get year, month lengths based on leap years
    if year in leap_years:
        year_length = 366
        month_lengths = days_in_months_leap
    else:
        year_length = 365
        month_lengths = days_in_months

    # get data for selected year
    data_P1_year = ERA5_P1_daily[year_counter:year_counter+year_length,:,:]

    # loop through months in selected year
    month_counter = 0
    for m in range(12):
        month_length = month_lengths[m]
        # get data for selected month
        data_P1_month = data_P1_year[month_counter:month_counter+month_length,:,:]
        # average / sum across month
        ERA5_P1_monthly[(y*12)+m,:,:] = np.sum(data_P1_month,axis=0)
        month_counter += month_length
    # end loop through months
    year_counter += year_length
# end loop through years

print(ERA5_P1_monthly.shape)
```

```python
### Precipitation 2 ###
filename_P2 = 'ERA5_Precip_1994_2006_hourly_ecmwfgrid.nc'
ERA5_P2 = Dataset(dir_data + '/' + filename_P2)

# Extract variables
ERA5_P2_time = ERA5_P2.variables['valid_time'][:]     # hours since 01/01/1900
ERA5_P2_hourly = ERA5_P2.variables['tp'][:]      # total precipitation (m)

## Convert to daily
# Sum all precip in each day
numperday = 24  # hourly data has 24 time points per day
ERA5_P2_daily = np.zeros((int(len(ERA5_P2_time)/numperday),len(ERA5_P_lat), len(ERA5_P_lon)))
for daynum in range(int(len(ERA5_P2_time)/numperday)):
    inds = list(range(daynum*numperday, daynum*numperday+numperday))
    ERA5_P2_daily[daynum,:,:] = np.sum(ERA5_P2_hourly[inds,:,:],axis=0)

print(ERA5_P2_daily.shape)

## Convert to monthly
years_subset2 = np.arange(1994,2007,1)
ERA5_P2_monthly = np.zeros((len(years_subset2)*12,len(ERA5_P_lat),len(ERA5_P_lon)))

# loop through years
year_counter = 0
for y, year in enumerate(years_subset2):
    # get year, month lengths based on leap years
    if year in leap_years:
        year_length = 366
        month_lengths = days_in_months_leap
    else:
        year_length = 365
        month_lengths = days_in_months

    # get data for selected year
    data_P2_year = ERA5_P2_daily[year_counter:year_counter+year_length,:,:]

    # loop through months in selected year
    month_counter = 0
    for m in range(12):
        month_length = month_lengths[m]
        # get data for selected month
        data_P2_month = data_P2_year[month_counter:month_counter+month_length,:,:]
        # average / sum across month
        ERA5_P2_monthly[(y*12)+m,:,:] = np.sum(data_P2_month,axis=0)
        month_counter += month_length
    # end loop through months
    year_counter += year_length
# end loop through years

print(ERA5_P2_monthly.shape)
```

```python
### Precipitation 3 ###
filename_P3 = 'ERA5_Precip_2007_2017_hourly_ecmwfgrid.nc'
ERA5_P3 = Dataset(dir_data + '/' + filename_P3)

# Extract variables
ERA5_P3_time = ERA5_P3.variables['valid_time'][:]     # hours since 01/01/1900
ERA5_P3_hourly = ERA5_P3.variables['tp'][:]      # total precipitation (m)

## Convert to daily
# Sum all precip in each day
numperday = 24  # hourly data has 24 time points per day
ERA5_P3_daily = np.zeros((int(len(ERA5_P3_time)/numperday),len(ERA5_P_lat), len(ERA5_P_lon)))
for daynum in range(int(len(ERA5_P3_time)/numperday)):
    inds = list(range(daynum*numperday, daynum*numperday+numperday))
    ERA5_P3_daily[daynum,:,:] = np.sum(ERA5_P3_hourly[inds,:,:],axis=0)

print(ERA5_P3_daily.shape)

## Convert to monthly
years_subset3 = np.arange(2007,2018,1)
ERA5_P3_monthly = np.zeros((len(years_subset3)*12,len(ERA5_P_lat),len(ERA5_P_lon)))

# loop through years
year_counter = 0
for y, year in enumerate(years_subset3):
    # get year, month lengths based on leap years
    if year in leap_years:
        year_length = 366
        month_lengths = days_in_months_leap
    else:
        year_length = 365
        month_lengths = days_in_months

    # get data for selected year
    data_P3_year = ERA5_P3_daily[year_counter:year_counter+year_length,:,:]

    # loop through months in selected year
    month_counter = 0
    for m in range(12):
        month_length = month_lengths[m]
        # get data for selected month
        data_P3_month = data_P3_year[month_counter:month_counter+month_length,:,:]
        # average / sum across month
        ERA5_P3_monthly[(y*12)+m,:,:] = np.sum(data_P3_month,axis=0)
        month_counter += month_length
    # end loop through months
    year_counter += year_length
# end loop through years

print(ERA5_P3_monthly.shape)
```

```python
### combine 3 precip datasets together

ERA5_P_monthly = np.zeros((len(years_all)*12,len(ERA5_P_lat),len(ERA5_P_lon)))
ERA5_P_monthly[0:len(years_subset1)*12,:,:] = ERA5_P1_monthly
ERA5_P_monthly[len(years_subset1)*12:len(years_subset1)*12+len(years_subset2)*12,:,:] = ERA5_P2_monthly
ERA5_P_monthly[len(years_subset1)*12+len(years_subset2)*12:,:,:] = ERA5_P3_monthly

print(ERA5_P_monthly.shape)
```

```python
### Extract grid points within Mica basin

## new array where all gridpoints not in basin set to 10
# to be masked later for plotting
# start with array full of 10s, then add in true values from gridboxes in basin
Tmax_basin = np.full(ERA5_Tmax_monthly.shape, 10, dtype=float)
Tmin_basin = np.full(ERA5_Tmin_monthly.shape, 10, dtype=float)
Tavg_basin = np.full(ERA5_Tavg_monthly.shape, 10, dtype=float)
P_basin = np.full(ERA5_P_monthly.shape, 10, dtype=float)

# loop through all lat,lons to determine if in basin
grid_counter = 0  # count number of gridboxes in basin
for lt, lat in enumerate(gridedgey):
    for ln, lon in enumerate(gridedgex):
        ss = polygon.contains(Point(UTM_lon[lt,ln], UTM_lat[lt,ln]))  # top left corner
        UTMlonxx, UTMlatxx = myProj(lon,lat+ystep)
        xx = polygon.contains(Point(UTMlonxx, UTMlatxx))  # bottom left corner
        UTMlonyy, UTMlatyy = myProj(lon+xstep,lat+ystep)
        yy = polygon.contains(Point(UTMlonyy, UTMlatyy)) # bottom right corner
        UTMlonzz, UTMlatzz = myProj(lon+xstep,lat)
        zz = polygon.contains(Point(UTMlonzz, UTMlatzz)) # top right corner
        UTMlonsx, UTMlatsx = myProj(lon,lat+ystep/2)
        sx = polygon.contains(Point(UTMlonsx, UTMlatsx)) # left mid point
        UTMlonzy, UTMlatzy = myProj(lon+xstep,lat+ystep/2)
        zy = polygon.contains(Point(UTMlonzy, UTMlatzy)) # right mid point
        UTMlonsz, UTMlatsz = myProj(lon+xstep/2,lat)
        sz = polygon.contains(Point(UTMlonsz, UTMlatsz)) # top mid point
        UTMlonxy, UTMlatxy = myProj(lon+xstep/2,lat+ystep)
        xy = polygon.contains(Point(UTMlonxy, UTMlatxy)) # bottom mid point
        UTMlonmid, UTMlatmid = myProj(lon+xstep/2,lat+ystep/2)
        mid = polygon.contains(Point(UTMlonmid, UTMlatmid)) # middle of gridbox

        if ss == True:  # if top left corner in basin shape
            #print("ss: Contains lat={},lon={}".format(lat,lon))
            Tmax_basin[:,lt,ln] = ERA5_Tmax_monthly[:,lt,ln] # add to placeholder arrays
            Tmin_basin[:,lt,ln] = ERA5_Tmin_monthly[:,lt,ln]
            Tavg_basin[:,lt,ln] = ERA5_Tavg_monthly[:,lt,ln]
            P_basin[:,lt,ln] = ERA5_P_monthly[:,lt,ln]
            grid_counter += 1
        elif xx == True:  # if bottom left corner in basin shape
            #print("xx: Contains lat={},lon={}".format(lat,lon))
            Tmax_basin[:,lt,ln] = ERA5_Tmax_monthly[:,lt,ln]
            Tmin_basin[:,lt,ln] = ERA5_Tmin_monthly[:,lt,ln]
            Tavg_basin[:,lt,ln] = ERA5_Tavg_monthly[:,lt,ln]
            P_basin[:,lt,ln] = ERA5_P_monthly[:,lt,ln]
            grid_counter += 1
        elif yy == True:  # if bottom right corner in basin shape
            #print("yy: Contains lat={},lon={}".format(lat,lon))
            Tmax_basin[:,lt,ln] = ERA5_Tmax_monthly[:,lt,ln]
            Tmin_basin[:,lt,ln] = ERA5_Tmin_monthly[:,lt,ln]
            Tavg_basin[:,lt,ln] = ERA5_Tavg_monthly[:,lt,ln]
            P_basin[:,lt,ln] = ERA5_P_monthly[:,lt,ln]
            grid_counter += 1
        elif zz == True:  # if top right corner in basin shape
            #print("zz: Contains lat={},lon={}".format(lat,lon))
            Tmax_basin[:,lt,ln] = ERA5_Tmax_monthly[:,lt,ln]
            Tmin_basin[:,lt,ln] = ERA5_Tmin_monthly[:,lt,ln]
            Tavg_basin[:,lt,ln] = ERA5_Tavg_monthly[:,lt,ln]
            P_basin[:,lt,ln] = ERA5_P_monthly[:,lt,ln]
            grid_counter += 1
        elif mid == True:  # if midpoint in basin shape
            #print("mid: Contains lat={},lon={}".format(lat,lon))
            Tmax_basin[:,lt,ln] = ERA5_Tmax_monthly[:,lt,ln]
            Tmin_basin[:,lt,ln] = ERA5_Tmin_monthly[:,lt,ln]
            Tavg_basin[:,lt,ln] = ERA5_Tavg_monthly[:,lt,ln]
            P_basin[:,lt,ln] = ERA5_P_monthly[:,lt,ln]
            grid_counter += 1
        elif sx == True and zy == True: # if left and right midpoints both in basin shape
            #print("sx,zy: Contains lat={},lon={}".format(lat,lon))
            Tmax_basin[:,lt,ln] = ERA5_Tmax_monthly[:,lt,ln]
            Tmin_basin[:,lt,ln] = ERA5_Tmin_monthly[:,lt,ln]
            Tavg_basin[:,lt,ln] = ERA5_Tavg_monthly[:,lt,ln]
            P_basin[:,lt,ln] = ERA5_P_monthly[:,lt,ln]
            grid_counter += 1
        elif sz == True and xy == True: # if top and bottom midpoints both in basin shape
            #print("sz,xy: Contains lat={},lon={}".format(lat,lon))
            Tmax_basin[:,lt,ln] = ERA5_Tmax_monthly[:,lt,ln]
            Tmin_basin[:,lt,ln] = ERA5_Tmin_monthly[:,lt,ln]
            Tavg_basin[:,lt,ln] = ERA5_Tavg_monthly[:,lt,ln]
            P_basin[:,lt,ln] = ERA5_P_monthly[:,lt,ln]
            grid_counter += 1
# end loop through lats, lons
print("Station: {}, # grid points: {}".format('Mica', grid_counter))
```

### 3. Monthly ECMWF SEAS5 Hindcasts Temperature & Precipitation

Download ECMWF SEAS5 hindcasts: Seasonal forecast monthly statistics on single levels (Johnson et al., 2019): https://cds.climate.copernicus.eu/datasets/seasonal-monthly-single-levels?tab=download
- Originating centre: ECMWF
- System: 5
- Variables: 'Maximum temperature in the last 24 hours', 'Minimum temperature in the last 24 hours' and 'Total precipitation'
- Product type: Individual members, monthly mean
- Years: 1981 to 2017 (inclusive)
- Months: January
- Leadtime month: 1-6 (all)
- Subregion extraction: I download the whole province of BC, latitudes 48 to 60 degN; longitudes -142 to -109 degE

Final dataset should be located in Data/ directory, called:

SEAS5_forecasts_1981_2017_monthly.nc

Will also need Mica basin catchment outline, in Data/ directory, called:

Mica_basin_outline.csv

The following code extracts the data for the Mica catchment area and saves as a .csv file.

```python
### load dataset

filename_SEAS5 = 'SEAS5_forecasts_1981_2017_monthly.nc'
SEAS5_data = Dataset(dir_data + '/' + filename_SEAS5)

# Extract variables
SEAS5_lon = SEAS5_data.variables['longitude'][:]  # longitude in degW
SEAS5_lat = SEAS5_data.variables['latitude'][:]   # latitude in degW
SEAS5_initdate = SEAS5_data.variables['forecast_reference_time'][:]     # datetime64[ns] (initialization date)
SEAS5_leadtime = SEAS5_data.variables['forecastMonth'][:]
SEAS5_Tmax = SEAS5_data.variables['mx2t24'][:]      # max temperature (K)
SEAS5_Tmin = SEAS5_data.variables['mn2t24'][:]      # min temperature (K)
SEAS5_P = SEAS5_data.variables['tprate'][:]      # precipitation (m)

print("Longitude shape = ", SEAS5_lon.shape)
print("Latitude shape = ", SEAS5_lat.shape)
print(SEAS5_Tmax.shape)  # (number, initialization_time, lead_month, latitude, longitude)
```

```python
### Reshape data to 1D in time
# old shape: (number, initialization_time, lead_month, latitude, longitude)
# new shape: (number, time, lat, lon)

# initialization years
init_years = np.arange(1981,2018,1)  # each has 6 months of forecast data

# placeholder arrays
SEAS5_Tmax_reshape = np.zeros((SEAS5_Tmax.shape[0], len(init_years)*6, SEAS5_Tmax.shape[3], SEAS5_Tmax.shape[4]))
SEAS5_Tmin_reshape = np.zeros((SEAS5_Tmin.shape[0], len(init_years)*6, SEAS5_Tmin.shape[3], SEAS5_Tmin.shape[4]))
SEAS5_P_reshape = np.zeros((SEAS5_P.shape[0], len(init_years)*6, SEAS5_P.shape[3], SEAS5_P.shape[4]))

# loop through initialization years
for y, year in enumerate(init_years):
    for m in range(6):
        SEAS5_Tmax_reshape[:,(y*6)+m,:,:] = SEAS5_Tmax[:,y,m,:,:]
        SEAS5_Tmin_reshape[:,(y*6)+m,:,:] = SEAS5_Tmin[:,y,m,:,:]
        SEAS5_P_reshape[:,(y*6)+m,:,:] = SEAS5_P[:,y,m,:,:]
# end loop through init years

print(SEAS5_Tmax_reshape.shape)
```

```python
### Average over ensemble members

# only 25 ensemble members 1981-2016 so average over first 25 only for all years

SEAS5_Tmax_ensavg = np.mean(SEAS5_Tmax_reshape[:25,:,:,:], axis=0)
SEAS5_Tmin_ensavg = np.mean(SEAS5_Tmin_reshape[:25,:,:,:], axis=0)
SEAS5_P_ensavg = np.mean(SEAS5_P_reshape[:25,:,:,:], axis=0)

print(SEAS5_Tmax_ensavg.shape)
```

```python
### Extract grid points within Mica basin

## new array where all gridpoints not in basin set to 10
# to be masked later for plotting
# start with array full of 10s, then add in true values from gridboxes in basin
Tmax_basin = np.full(SEAS5_Tmax_ensavg.shape, 10, dtype=float)
Tmin_basin = np.full(SEAS5_Tmin_ensavg.shape, 10, dtype=float)
P_basin = np.full(SEAS5_P_ensavg.shape, 10, dtype=float)

# loop through all lat,lons to determine if in basin
grid_counter = 0  # count number of gridboxes in basin
for lt, lat in enumerate(gridedgey):
    for ln, lon in enumerate(gridedgex):
        ss = polygon.contains(Point(UTM_lon[lt,ln], UTM_lat[lt,ln]))  # top left corner
        UTMlonxx, UTMlatxx = myProj(lon,lat+ystep)
        xx = polygon.contains(Point(UTMlonxx, UTMlatxx))  # bottom left corner
        UTMlonyy, UTMlatyy = myProj(lon+xstep,lat+ystep)
        yy = polygon.contains(Point(UTMlonyy, UTMlatyy)) # bottom right corner
        UTMlonzz, UTMlatzz = myProj(lon+xstep,lat)
        zz = polygon.contains(Point(UTMlonzz, UTMlatzz)) # top right corner
        UTMlonsx, UTMlatsx = myProj(lon,lat+ystep/2)
        sx = polygon.contains(Point(UTMlonsx, UTMlatsx)) # left mid point
        UTMlonzy, UTMlatzy = myProj(lon+xstep,lat+ystep/2)
        zy = polygon.contains(Point(UTMlonzy, UTMlatzy)) # right mid point
        UTMlonsz, UTMlatsz = myProj(lon+xstep/2,lat)
        sz = polygon.contains(Point(UTMlonsz, UTMlatsz)) # top mid point
        UTMlonxy, UTMlatxy = myProj(lon+xstep/2,lat+ystep)
        xy = polygon.contains(Point(UTMlonxy, UTMlatxy)) # bottom mid point
        UTMlonmid, UTMlatmid = myProj(lon+xstep/2,lat+ystep/2)
        mid = polygon.contains(Point(UTMlonmid, UTMlatmid)) # middle of gridbox

        if ss == True:  # if top left corner in basin shape
            #print("ss: Contains lat={},lon={}".format(lat,lon))
            Tmax_basin[:,lt,ln] = SEAS5_Tmax_ensavg[:,lt,ln] # add to placeholder arrays
            Tmin_basin[:,lt,ln] = SEAS5_Tmin_ensavg[:,lt,ln]
            P_basin[:,lt,ln] = SEAS5_P_ensavg[:,lt,ln]
            grid_counter += 1
        elif xx == True:  # if bottom left corner in basin shape
            #print("xx: Contains lat={},lon={}".format(lat,lon))
            Tmax_basin[:,lt,ln] = SEAS5_Tmax_ensavg[:,lt,ln]
            Tmin_basin[:,lt,ln] = SEAS5_Tmin_ensavg[:,lt,ln]
            P_basin[:,lt,ln] = SEAS5_P_ensavg[:,lt,ln]
            grid_counter += 1
        elif yy == True:  # if bottom right corner in basin shape
            #print("yy: Contains lat={},lon={}".format(lat,lon))
            Tmax_basin[:,lt,ln] = SEAS5_Tmax_ensavg[:,lt,ln]
            Tmin_basin[:,lt,ln] = SEAS5_Tmin_ensavg[:,lt,ln]
            P_basin[:,lt,ln] = SEAS5_P_ensavg[:,lt,ln]
            grid_counter += 1
        elif zz == True:  # if top right corner in basin shape
            #print("zz: Contains lat={},lon={}".format(lat,lon))
            Tmax_basin[:,lt,ln] = SEAS5_Tmax_ensavg[:,lt,ln]
            Tmin_basin[:,lt,ln] = SEAS5_Tmin_ensavg[:,lt,ln]
            P_basin[:,lt,ln] = SEAS5_P_ensavg[:,lt,ln]
            grid_counter += 1
        elif mid == True:  # if midpoint in basin shape
            #print("mid: Contains lat={},lon={}".format(lat,lon))
            Tmax_basin[:,lt,ln] = SEAS5_Tmax_ensavg[:,lt,ln]
            Tmin_basin[:,lt,ln] = SEAS5_Tmin_ensavg[:,lt,ln]
            P_basin[:,lt,ln] = SEAS5_P_ensavg[:,lt,ln]
            grid_counter += 1
        elif sx == True and zy == True: # if left and right midpoints both in basin shape
            #print("sx,zy: Contains lat={},lon={}".format(lat,lon))
            Tmax_basin[:,lt,ln] = SEAS5_Tmax_ensavg[:,lt,ln]
            Tmin_basin[:,lt,ln] = SEAS5_Tmin_ensavg[:,lt,ln]
            P_basin[:,lt,ln] = SEAS5_P_ensavg[:,lt,ln]
            grid_counter += 1
        elif sz == True and xy == True: # if top and bottom midpoints both in basin shape
            #print("sz,xy: Contains lat={},lon={}".format(lat,lon))
            Tmax_basin[:,lt,ln] = SEAS5_Tmax_ensavg[:,lt,ln]
            Tmin_basin[:,lt,ln] = SEAS5_Tmin_ensavg[:,lt,ln]
            P_basin[:,lt,ln] = SEAS5_P_ensavg[:,lt,ln]
            grid_counter += 1
# end loop through lats, lons

print("Station: {}, # grid points: {}".format('Mica', grid_counter))
```

```python
### Average over all gridpoints in basin

# placeholders for gridpoints in basin
Tmax_basin_only = np.zeros((SEAS5_Tmax_ensavg.shape[0],grid_counter), dtype=float)
Tmin_basin_only = np.zeros((SEAS5_Tmin_ensavg.shape[0],grid_counter), dtype=float)
P_basin_only = np.zeros((SEAS5_P_ensavg.shape[0],grid_counter), dtype=float)

# get indices of points in basin
inds_basin = np.where(Tmax_basin[0,:,:] > 10)
inds_basin0 = inds_basin[0]
inds_basin1 = inds_basin[1]
# loop through gridpoints in basin:
for ii in range(len(inds_basin0)):
    Tmax_basin_only[:,ii] = Tmax_basin[:,inds_basin0[ii],inds_basin1[ii]]
    Tmin_basin_only[:,ii] = Tmin_basin[:,inds_basin0[ii], inds_basin1[ii]]
    P_basin_only[:,ii] = P_basin[:,inds_basin0[ii], inds_basin1[ii]]

# average over all grid points in basin
Tmax_basin_only_avg = np.mean(Tmax_basin_only, axis=1)
Tmin_basin_only_avg = np.mean(Tmin_basin_only, axis=1)
P_basin_only_avg = np.mean(P_basin_only, axis=1)
```

```python
### convert precip rate (m/s) to total precip (m)
# multiply by total seconds in month

Paccum_basin_only_avg = np.empty_like(P_basin_only_avg)

leap_years = [1980, 1984, 1988, 1992, 1996, 2000, 2004, 2008, 2012, 2016, 2020]
month_lengths = np.array([31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31])  # days
month_lengths_leap = np.array([31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31])

# convert month_lengths from days to seconds
month_seconds = month_lengths*24*3600  # seconds
month_seconds_leap = month_lengths_leap*24*3600

# loop through initialization years
counter = 0
for y, year in enumerate(init_years):
    if year in leap_years:
        times = month_seconds_leap
    else:
        times = month_seconds
    # loop through months
    for n in range(6):
        Paccum_basin_only_avg[counter] = P_basin_only_avg[counter]*times[n]
        counter += 1
# end loop through initialization years
```

```python
### plot precip rate and total precip

fig = plt.figure(figsize=(10,3))
plt.plot(P_basin_only_avg[:6], label='Precip_rate')
plt.xlabel("Months")
plt.ylabel("Precipitation rate (m/s)")
plt.legend()
plt.show()

fig = plt.figure(figsize=(10,3))
plt.plot(Paccum_basin_only_avg[:6], label='Precip_accumulated')
plt.xlabel("Months")
plt.ylabel("Total Precipitation (m)")
plt.legend()
plt.show()
```

```python
### save to csv file for easy access later

## first, create array of dates
dates_1D = np.zeros(SEAS5_Tmax_ensavg.shape[0], dtype='U100')
for y, year in enumerate(init_years):
    for m in range(6):
        dates_1D[(y*6)+m] = '0' + str(m+1) + '-' + str(year)

## save to csv
with open(dir_data + '/SEAS5_monthly_input_Mica_1981_2017.csv','w',newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Date', 'Min Temp (K)', 'Max Temp (K)', 'Precip (m)'])
    for i in range(len(dates_1D)):
        row_data = [dates_1D[i], Tmin_basin_only_avg[i], Tmax_basin_only_avg[i], Paccum_basin_only_avg[i]]
        writer.writerow(row_data)
```
